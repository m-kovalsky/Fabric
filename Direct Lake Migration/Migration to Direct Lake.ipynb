## Create the Direct Lake model based on the import/DQ semantic model

#### *Calculated columns and calculated tables are not migrated to the Direct Lake model as they are not supported in Direct Lake mode*

```python
import time
import fabric_cat_tools as fct

newDatasetName = '' #Enter the new Direct Lake semantic model name

fct.create_blank_semantic_model(newDatasetName)

time.sleep(5)

fct.migrate_calc_tables_to_lakehouse(datasetName, newDatasetName)
fct.migrate_field_parameters(datasetName, newDatasetName)
fct.migrate_tables_columns_to_semantic_model(datasetName, newDatasetName)
fct.migrate_calc_tables_to_semantic_model(datasetName, newDatasetName)
fct.migrate_model_objects_to_semantic_model(datasetName, newDatasetName)
fct.refresh_semantic_model(newDatasetName)
fct.refresh_calc_tables(newDatasetName)
fct.refresh_semantic_model(newDatasetName)
```

## Rebind all reports using the old semantic model to the new Direct Lake semantic model

```python
import fabric_cat_tools as fct

fct.report_rebind_all(datasetName, newDatasetName)
```

## Rebind reports one-by-one (optional)

```python
import fabric_cat_tools as fct

reportName = '' # Enter report name which you want to rebind to the new Direct Lake model

fct.report_rebind(reportName, newDatasetName)
```
